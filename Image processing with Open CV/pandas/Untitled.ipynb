```python
#dataFrame oluşturma

boy= np.random.randint(150,190,(10))
kilo = np.random.randint(150,190,(10))
ogr_no= np.random.randint(2100,2200,(10))

my_dick= {"ogrenci":ogr_no,"ogr_boy":boy,"ogr_kilo":kilo}
my_df=pd.DataFrame.from_dict(my_dick)
my_df
```

```python
vize_not=np.random.randint(10,100,(10))
final_not=np.random.randint(10,100,(10))

my_df["ogr_vize"]=vize_not
my_df["ogr_final"]=final_not
my_df
```

```python
ogr_geçti=np.random.randint(10,100,(10))

ogr_geçti = (vize_not*40/100)+(final_not*60/100)
my_df["geçti_kaldı"]=ogr_geçti
my_df
```

```python
my_df["Durum"] = my_df["geçti_kaldı"]>50
my_df
```

```python
st_kilo=my_df["ogr_kilo"]>70
st_boy=my_df["ogr_boy"]>175

print(st_kilo)
print(st_boy)
```
