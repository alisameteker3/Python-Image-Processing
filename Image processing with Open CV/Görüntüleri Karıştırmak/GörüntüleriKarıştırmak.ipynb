```python
import cv2
import matplotlib.pylab as plt
```

```python
img1 = cv2.imread("./manzara.jpg")
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread("./denizkiyisi.jpg")
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

plt.figure()
plt.imshow(img1)

plt.figure()
plt.imshow(img2)
```

```python
print(img1.shape)
print(img2.shape)
```

```python
img1 = cv2.resize(img1 , (450,450))
print(img1.shape)

img2 = cv2.resize(img2,(450,450))
print(img2.shape)
```

```python
plt.figure()
plt.imshow(img1)

plt.figure()
plt.imshow(img2)
```

```python
blended = cv2.addWeighted(src1 = img1, alpha = 0.5 , src2 = img2 , beta = 0.5 ,gamma = 0)
plt.figure()
plt.imshow(blended)
```
